### 標準偏差


$$
標準偏差=\sigma=\sqrt{\frac{1}{n} \sum_{i=1}^{n}\left(x_{i}-\mu\right)^{2}}
$$


 $$
分散=\sigma^{2}
$$
 　
 標本の散らばり具合を定量化したもの（参考： 偏差値）

### 共分散行列
　
 $$
S \mathbf{x y}=\frac{1}{n}(\mathbf{x}-\overline{\mathbf{x}})^{\prime}(\mathbf{y}-\overline{\mathbf{y}})
$$

$$
\mathbf{S}=\left[\begin{array}{lll}{S x x} & {S x y} & {S x z} \\ {S x y} & {S y y} & {S y z} \\ {S x z} & {S y z} & {S z z}\end{array}\right]
$$

イメージは変数が複数あるときの、関係性をしめした行列（共分散、相関係数）


### 1次元の正規分布と多次元の正規分布
 

1次元の正規分布　　
独立な多数の因子の和として表される確率変数は正規分布に従う


 $$
f(x)=\frac{1}{\sqrt{2 \pi \sigma^{2}}} \exp \left(-\frac{(x-\mu)^{2}}{2 \sigma^{2}}\right) \quad(x \in \mathbb{R})
$$



多次元の正規分布

$$
f(x)=\frac{1}{(\sqrt{2 \pi})^{n} \sqrt{|\Sigma|}} \exp \left(-\frac{1}{2}(x-\mu)^{\mathrm{T}} \Sigma^{-1}(x-\mu)\right)
$$

```python
def mean_cov_plot(mu1=[-3, 0], mu2=[0, -3], sigma=[[1.0, 0.8], [0.8, 1.0]]):
    
    # mu1 = [-3, 0]
    # mu2 = [0, -3]
    # sigma = [[1.0, 0.8], [0.8, 1.0]]
    
    data1 = multivariate_normal(mu1, sigma, 500)
    data2 = multivariate_normal(mu2, sigma, 500)
    
    data_all = np.concatenate([data1, data2])
    data_all_labelled = np.insert(data_all, 2, label, axis=1)

    x = data_all_labelled[:, 0]
    y = data_all_labelled[:, 1]

    plt.scatter(x, y, s=50, c=data_all_labelled[:, 2], marker='o')
    plt.show()
    return
```

```python
mean_cov_plot()
```

```python
mean_cov_plot([3, 0], [0, -3], [[1.0, 0.8], [0.8, 1.0]])
```
